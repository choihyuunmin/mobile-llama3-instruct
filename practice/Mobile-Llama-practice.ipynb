```python
import os
from typing import List


import torch
from torch import cuda, bfloat16
import matplotlib.pyplot as plt
import matplotlib as mpl

import transformers
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    pipeline,
    logging,
)

from pylab import rcParams
```

```python
# 로컬용 환경변수 load
from dotenv import load_dotenv
load_dotenv()
```

```python
model_id = 'hyonbokan/mobile_llama_5kRounds'

# mac os 설정 확인
if torch.cuda.is_available():
    device = f'cuda:{torch.cuda.current_device()}'
elif torch.backends.mps.is_available():
    device = torch.device('mps')
else:
    device = 'cpu'
```

```python
# huggingface token
hf_auth = os.environ.get('hf_token')

# model config
model_config = transformers.AutoConfig.from_pretrained(
    model_id,
    use_auth_token=hf_auth
)

bnb_config = transformers.BitsAndBytesConfig(
    load_in_8bit=True,
)
```

```python
# model load
model = transformers.AutoModelForCausalLM.from_pretrained(
    model_id,
    trust_remote_code=True,
    config=model_config,
    quantization_config=bnb_config,
    device_map='auto',
    use_auth_token=hf_auth
)
model.eval()
print(f"Model loaded on {device}")
```

```python
# tokenizer load
tokenizer = transformers.AutoTokenizer.from_pretrained(
    model_id,
    use_auth_token=hf_auth
)

# tokenizer.pad_token_id = (0)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"
```

```python
# chat with llm
prompt = "Use PyBGPStream to perform BGP analysis and identify anomalies in AS path lengths for IPv4 prefixes during two different time windows: from March 5, 2023, 14:00 to 16:00, and from March 7, 2023, 14:00 to 16:00."
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=718)
result = pipe(f"<s>[INST] {prompt} [/INST]")
print(result[0]['generated_text'])
```

```python
# performance analysis
prompt = "Generate Python code that calculates Key Performance Indicators (KPIs) for 5G network performance including Total Network Capacity, Capacity per Area, Capacity per Point, Cost per Capacity, Cost per Area, and Surplus per Area. The code should also create visual plots for each of these KPIs and create a new DataFrame to store KPI data."
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=718)
result = pipe(f"<s>[INST] {prompt} [/INST]")
print(result[0]['generated_text'])
```

```python
# packet analysis
prompt = "Given a n3.pcap file containing network traffic data, create a Python script to process the PCAP. The script should extract the source IP, destination IP, source port, destination port, and the raw payload data for each packet in the PCAP. Store this data in a pandas DataFrame and display it."
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=718)
result = pipe(f"<s>[INST] {prompt} [/INST]")
print(result[0]['generated_text'])
```

```python
import sys
import pandas as pd
from scapy.all import rdpcap

# Specify the PCAP file path
pcap_file = './smallFlows.pcap'

try:
    packets = rdpcap(pcap_file)  # Read the PCAP file

    # Create a list to store the packet data
    packet_data = []

    for packet in packets:
        src_ip, dst_ip, src_port, dst_port, payload = None, None, None, None, None

        if packet.haslayer('IP'):
            src_ip = packet['IP'].src
            dst_ip = packet['IP'].dst

        if packet.haslayer('TCP'):
            src_port = packet['TCP'].sport
            dst_port = packet['TCP'].dport
        elif packet.haslayer('UDP'):
            src_port = packet['UDP'].sport
            dst_port = packet['UDP'].dport

        payload = bytes(packet.payload)

        packet_data.append([src_ip, dst_ip, src_port, dst_port, payload])

    # Create a pandas DataFrame
    df = pd.DataFrame(packet_data, columns=['Source IP', 'Destination IP', 'Source Port', 'Destination Port', 'Raw Payload'])

    # Display the DataFrame
    print(df)

except Exception as e:
    exc_type, exc_obj, exc_tb = sys.exc_info()
    print(f"An error occurred: {e}")
    print(f"Error Type: {exc_type}")
    print(f"File: {exc_tb.tb_frame.f_code.co_filename}")
    print(f"Line: {exc_tb.tb_lineno}")
```
